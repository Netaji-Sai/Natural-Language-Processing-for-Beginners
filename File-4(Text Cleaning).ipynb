# **Text Cleaning Techniques**

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from collections import Counter
from nltk import pos_tag

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')

def split_sentences(content):
    return sent_tokenize(content)

def tokenize_and_stem(content):
    tokens = word_tokenize(content)
    stemmer = PorterStemmer()
    stem_tokens = [stemmer.stem(token) for token in tokens]
    return tokens, stem_tokens

def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    text_excluding_stop_words = [token for token in tokens if token.lower() not in stop_words]
    return text_excluding_stop_words

def remove_rare_words(tokens, rare_threshold=1):
    word_frequencies = Counter(tokens)
    text_excluding_rare_words = [token for token in tokens if word_frequencies[token] > rare_threshold]
    return text_excluding_rare_words

def display_parts_of_speech(tokens):
    postag_word = pos_tag(tokens)
    return postag_word

with open("Example.txt", "r") as f:
    content = f.read()

# 1. Split Sentences
sentences = split_sentences(content)
print("Split Sentences:")
print(sentences)

# 2. Tokenization and Stemming
tokens, stem_tokens = tokenize_and_stem(content)
print("Tokenization and Stemming:")
print("Tokens:", tokens)
print("Stemmed Tokens:", stem_tokens)

# 3. Remove Stopwords and Rare Words
text_excluding_stop_words = remove_stopwords(tokens)
print("Remove Stopwords:")
print(text_excluding_stop_words)

text_excluding_rare_words = remove_rare_words(tokens)
print("Remove Rare Words:")
print(text_excluding_rare_words)

# 4. Display Parts of Speech
postag_word = display_parts_of_speech(tokens)
print("Display Parts of Speech:")
print(postag_word)
```

```python
#SPlitting the text to sentences
sentences = sent_tokenize(content)
print(sentences)
```

```python
#SPlitting into words
words = word_tokenize(content)

print(words)
print(len(words))
```

```python
#Word counter
freq = nltk.FreqDist(words)
print(freq)
for k,v in freq.items():
  print(k," : ",v)
```

```python
#Stemmming
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(token) for token in words]
for i in range(0,len(stemmed_words)):
  print(words[i]," : ",stemmed_words[i])
```

```python
#Stop Word Removal
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
print(stop_words)

after_removal_stopwords = [token for token in words if token not in stop_words]
print(f'Non Stop Words Count : {len(after_removal_stopwords)}')
print(after_removal_stopwords)
```

```python
#Rare Word Removal
from collections import Counter

word_frequencies = Counter(words)
rare_removal  = [token for token in words if word_frequencies[token]>1]
print(len(rare_removal))
print(rare_removal)
```

```python
#POS Tagging
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
pos_tagged_words = pos_tag(words)
print(pos_tagged_words)
```
